```python
# 're' stands for regular expressions
import re
from urlextract import URLExtract
import pandas as pd
import numpy as np
```

```python
f = open('WhatsApp Chat with Unofficial Cs(ai).txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = "\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s\w+\s-\s"
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message' : messages, 'message_date' : dates})

# a function to parse the date with the correct format
def parse_date(date_str):
    return pd.to_datetime(date_str.strip(), format='%m/%d/%y, %I:%M\u202F%p -')

# Convert message_date type
df['message_date'] = df['message_date'].apply(parse_date)

df.rename(columns={'message_date' : 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
# Separate users and messages
users = []
messages = []

for i in df['user_message']:
    entry = re.split('([\w\W]+?):\s',i)
    if entry[1:]:  # username and message exist
        users.append(entry[1])
        messages.append(entry[2])
    else:  # system message or notification
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
          if hour == 23:
                  period.append(str(hour) + "-" + str('00'))
          elif hour == 0:
                  period.append(str('00') + "-" + str(hour+1))
          else:
                  period.append(str(hour) + "-" + str(hour+1))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_analysis(selected_user, df):
    if selected_user != "Overall":
        df = df[df["user"] == selected_user]

    analyzer = SentimentIntensityAnalyzer()
    df['sentiment'] = df['message'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

    # Calculate average sentiment score
    avg_sentiment = df['sentiment'].mean()

    return avg_sentiment, df['sentiment']
```
